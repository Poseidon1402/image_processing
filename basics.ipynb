## Import the necessary libraries

```python
import cv2
import numpy as np
from PIL import Image, ImageOps, ImageDraw
import matplotlib.pyplot as plt
```

## Read images

```python
# Load image with PIL
img = Image.open('dog.jpg')

# Load image using OpenCV
img = cv2.imread('dog.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB

# display loaded image with matplotlib
# Same result whatever the loading method
plt.imshow(img)
```

## Manipulating images

```python
# Flip image vertically
img = Image.open('dog.jpg')
img_flipped = ImageOps.flip(img)

# Display images side by side
plt.subplot(1, 2, 1)
plt.imshow(img_flipped)
plt.title('Flipped Image')
# Rotate image by 45 degrees
img_rotated = img.rotate(45)
plt.subplot(1, 2, 2)
plt.title('Rotated Image')
plt.imshow(img_rotated)
# Mirror image horizontally
img_mirrored = ImageOps.mirror(img)
plt.figure()
plt.imshow(img_mirrored)
plt.title('Mirrored Image')
```

```python
# Rotate image 90 degrees clockwise using OpenCV
img = cv2.imread('dog.jpg')

img_flip = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)

# Display images side by side
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(img_flip, cv2.COLOR_BGR2RGB))
plt.title('Rotated Image')
```

## Draw shape

```python
img = Image.open('dog.jpg')

draw = img.copy()
img_fn = ImageDraw.Draw(im=draw)

LEFT = 200
UPPER = 200
RIGHT = 600
LOWER = 600
shape = [LEFT, UPPER, RIGHT, LOWER]
img_fn.rectangle(xy=shape, fill="red")

plt.imshow(draw)
```

```python
img = cv2.imread('dog.jpg')
img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
img_draw = np.copy(img)

LEFT = 200
UPPER = 200
RIGHT = 600
LOWER = 600

start_pt, end_pt = (LEFT, UPPER),(RIGHT, LOWER)

cv2.rectangle(img_draw, pt1=start_pt, pt2=end_pt, color=(0, 255, 0), thickness=3)

plt.imshow(img_draw)
```

## Pixel transformations

```python
img = cv2.imread('dog.jpg', cv2.IMREAD_GRAYSCALE)

# Thresholding
max_value = 255

# Use Otsu's binarization
ret,new_image = cv2.threshold(img, 0, 255, cv2.THRESH_OTSU)

# Display images side by side
plt.imshow(new_image)
```

# Scale image

```python
from PIL import Image

image = Image.open('lenna.png')

width = 512
height = 512

new_width = 2 * width
new_height = height

new_image = image.resize((new_width, new_height))

new_image
```

# Image rotation (PIL)

```python
from PIL import Image

image = Image.open("lenna.png")

theta = 45 # rotation angle in degrees

new_image = image.rotate(theta)

new_image
```
